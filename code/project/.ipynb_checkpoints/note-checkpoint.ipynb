```python
import os
import numpy as np
import h5py
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (14, 7)


print("[INFO] reading dataset....")
with h5py.File("/Users/saeedkazemi/Documents/Python/EE6563/Dataset/project/Step Scan Dataset/[H5]/footpressures_align.h5", "r") as hdf:
    barefoots = hdf.get("/barefoot/data")[:]
```

```python
import pandas as pd 
img = np.squeeze(barefoots[1, :, :, :])
i = plt.imshow(img[32, :, :])

data = list()
for sample in range(barefoots.shape[0]):
    img = np.squeeze(barefoots[sample, :, :, :])
    aa = np.sum(img, axis=2)
    bb = np.sum(aa, axis=1)
    data.append(bb[:60])

a = pd.DataFrame(data)#, columns="Sample_"+str(sample))
index = ["Sample_"+str(i) for i in np.arange(barefoots.shape[0]) ]
column = [i for i in np.arange(60) ]
df_totalpressure = pd.DataFrame(data, columns=column, index=index).T

df_totalpressure.to_pickle("/Users/saeedkazemi/Documents/Python/EE6563/Dataset/project/df_totalpressure.pickle")

```
